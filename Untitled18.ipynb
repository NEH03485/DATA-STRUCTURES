```python
# 1. Create a string with your name and print it
name = "Neha Evane"
print(name)

# 2. Find the length of the string "Hello World"
greeting = "Hello World"
print(len(greeting))

# 3. Slice the first 3 characters from the string "Python Programming"
text = "Python Programming"
print(text[:3])

# 4. Convert the string "hello" to uppercase
word = "hello"
print(word.upper())

# 5. Replace "apple" with "orange" in the string "I like apple"
sentence = "I like apple"
print(sentence.replace("apple", "orange"))

# 6. Create a list with numbers 1 to 5 and print it
numbers = [1, 2, 3, 4, 5]
print(numbers)

# 7. Append the number 10 to the list [1, 2, 3, 4]
my_list = [1, 2, 3, 4]
my_list.append(10)
print(my_list)

# 8. Remove the number 3 from the list [1, 2, 3, 4, 5]
num_list = [1, 2, 3, 4, 5]
num_list.remove(3)
print(num_list)

# 9. Access the second element in the list ['a', 'b', 'c', 'd']
letters = ['a', 'b', 'c', 'd']
print(letters[1])  # Index starts from 0

# 10. Reverse the list [10, 20, 30, 40, 50]
my_nums = [10, 20, 30, 40, 50]
my_nums.reverse()
print(my_nums)

# 11. Create a tuple with elements 100, 200, 300 and print it
my_tuple = (100, 200, 300)
print(my_tuple)

# 12. Access the second-to-last element from the tuple ('red', 'green', 'blue', 'yellow')
colors = ('red', 'green', 'blue', 'yellow')
print(colors[-2])
```

```python
# 13. Find the minimum number in the tuple (10, 20, 5, 15)
my_tuple = (10, 20, 5, 15)
print(min(my_tuple))

# 14. Find the index of the element "cat" in the tuple ('dog', 'cat', 'rabbit')
animals = ('dog', 'cat', 'rabbit')
print(animals.index('cat'))

# 15. Create a tuple of three fruits and check if "kiwi" is in it
fruits = ('apple', 'banana', 'grape')
print('kiwi' in fruits)

# 16. Create a set with elements 'a', 'b', 'c' and print it
letters = {'a', 'b', 'c'}
print(letters)

# 17. Clear all elements from the set {1, 2, 3, 4, 5}
nums = {1, 2, 3, 4, 5}
nums.clear()
print(nums)

# 18. Remove the element 4 from the set {1, 2, 3, 4}
s = {1, 2, 3, 4}
s.remove(4)
print(s)

# 19. Find the union of two sets {1, 2, 3} and {3, 4, 5}
set1 = {1, 2, 3}
set2 = {3, 4, 5}
print(set1.union(set2))

# 20. Find the intersection of two sets {1, 2, 3} and {2, 3, 4}
print(set1.intersection(set2))

# 21. Create a dictionary with keys "name", "age", and "city"
person = {"name": "Neha", "age": 21, "city": "Delhi"}
print(person)

# 22. Add new key-value pair "country": "USA"
person = {'name': 'John', 'age': 25}
person["country"] = "USA"
print(person)

# 23. Access the value for key "name"
data = {'name': 'Alice', 'age': 30}
print(data['name'])

# 24. Remove the key "age"
info = {'name': 'Bob', 'age': 22, 'city': 'New York'}
del info['age']
print(info)

# 25. Check if key "city" exists
profile = {'name': 'Alice', 'city': 'Paris'}
print('city' in profile)

# 26. Create a list, a tuple, and a dictionary and print them
my_list = [1, 2, 3]
my_tuple = (4, 5, 6)
my_dict = {'a': 1, 'b': 2}
print(my_list)
print(my_tuple)
print(my_dict)

# 27. Create list of 5 random numbers and sort in ascending order
import random
rand_nums = random.sample(range(1, 101), 5)
rand_nums.sort()
print(rand_nums)

# 28. Create a list with strings and print the element at index 3
words = ['apple', 'banana', 'cherry', 'date', 'kiwi']
print(words[3])

# 29. Combine two dictionaries and print the result
d1 = {'a': 1, 'b': 2}
d2 = {'c': 3, 'd': 4}
combined = {**d1, **d2}
print(combined)

# 30. Convert a list of strings into a set
string_list = ['apple', 'banana', 'apple', 'kiwi']
string_set = set(string_list)
print(string_set)
```
